```python
import os, json, warnings, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import joblib
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    average_precision_score, roc_auc_score,
    f1_score, fbeta_score,
    precision_score, recall_score,
    classification_report, confusion_matrix,
    precision_recall_curve
)
from tqdm.auto import tqdm
import optuna
optuna.logging.set_verbosity(optuna.logging.WARNING)
warnings.filterwarnings('ignore')
from google.colab import drive
drive.mount('/content/drive')

FILE_PATH = '/content/drive/MyDrive/ml/Final_Topology_Dataset_FULL_0_to_5__2.csv'
SAVE_DIR  = '/content/drive/MyDrive/alibaba_ms/saved_models/gb/'
os.makedirs(SAVE_DIR, exist_ok=True)

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device         : {DEVICE}")
print(f"Save directory : {SAVE_DIR}")
if DEVICE.type == 'cuda':
    print(f"GPU            : {torch.cuda.get_device_name(0)}")
else:
    print("NOTE           : No GPU found. Running on CPU.")
    print("               : Enable GPU in Runtime > Change Runtime Type")


# ─────────────────────────────────────────────────────────────
# BLOCK 1: LOAD DATASET
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 1: LOADING DATASET")
print("=" * 60)

t0 = time.time()
df = pd.read_csv(FILE_PATH)
df = df.sort_values('time_window').reset_index(drop=True)

NON_FEATURE_COLS = ['um', 'dm', 'time_window', 'violation_next_window']
FEATURE_COLS     = [c for c in df.columns if c not in NON_FEATURE_COLS]
TARGET           = 'violation_next_window'

X = df[FEATURE_COLS].fillna(-1).values.astype(np.float32)
y = df[TARGET].values.astype(np.float32)

print(f"  Loaded in      : {time.time()-t0:.1f}s")
print(f"  Total samples  : {len(df):,}")
print(f"  Features       : {len(FEATURE_COLS)}")
print(f"  Violation rate : {y.mean()*100:.2f}%")
print(f"  Normal         : {int((y==0).sum()):,}")
print(f"  Violated       : {int((y==1).sum()):,}")


# ─────────────────────────────────────────────────────────────
# BLOCK 2: TEMPORAL SPLITS
#
# Two sets of splits are prepared.
#
# (a) Subset — middle 30% of the dataset (idx 40%->70%).
#     Used exclusively for Optuna hyperparameter search.
#     This window is away from the cold-start first 40% and
#     the distribution-shift-prone last 30%, ensuring the
#     search reflects stable, representative conditions.
#     Internal split: 70% train | 30% val (no subset test —
#     final evaluation is performed on the full held-out set).
#
# (b) Full dataset — entire trace in temporal order.
#     Used for final model training, threshold optimisation,
#     cross-validation, and all reported evaluation metrics.
#     Split: 80% of first 95% as train | remainder as val
#            | last 5% as test.
#
# All splits are strictly temporal — no shuffling at any stage.
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 2: TEMPORAL SPLITS")
print("=" * 60)

n = len(df)

# Subset splits (Optuna search only)
sub_start     = int(n * 0.40)
sub_end       = int(n * 0.70)
X_sub         = X[sub_start:sub_end]
y_sub         = y[sub_start:sub_end]
n_sub         = len(X_sub)
sub_val_start = int(n_sub * 0.70)

X_sub_tr_raw = X_sub[:sub_val_start];  y_sub_tr = y_sub[:sub_val_start]
X_sub_vl_raw = X_sub[sub_val_start:];  y_sub_vl = y_sub[sub_val_start:]

assert y_sub_tr.sum() > 0, "No violations in subset train — adjust window"
assert y_sub_vl.sum() > 0, "No violations in subset val   — adjust window"

print(f"  Full dataset     : {n:,} samples")
print(f"  Subset window    : idx {sub_start}->{sub_end} ({n_sub:,} samples, middle 30%)")
print(f"  Sub train  : {len(X_sub_tr_raw):,} | Violations: {y_sub_tr.sum():.0f} ({y_sub_tr.mean()*100:.2f}%)")
print(f"  Sub val    : {len(X_sub_vl_raw):,}  | Violations: {y_sub_vl.sum():.0f} ({y_sub_vl.mean()*100:.2f}%)")

# Full splits (final model)
test_start_f = int(n * 0.95)
val_start_f  = int(n * 0.95 * 0.80)

X_tr_raw = X[:val_start_f];             y_tr = y[:val_start_f]
X_vl_raw = X[val_start_f:test_start_f]; y_vl = y[val_start_f:test_start_f]
X_te_raw = X[test_start_f:];            y_te = y[test_start_f:]

print(f"\n  Full train : {len(X_tr_raw):,} | Violations: {y_tr.sum():.0f} ({y_tr.mean()*100:.2f}%)")
print(f"  Full val   : {len(X_vl_raw):,}  | Violations: {y_vl.sum():.0f} ({y_vl.mean()*100:.2f}%)")
print(f"  Full test  : {len(X_te_raw):,}  | Violations: {y_te.sum():.0f} ({y_te.mean()*100:.2f}%)")


# ─────────────────────────────────────────────────────────────
# BLOCK 3: FEATURE SCALING
#
# Two independent scalers:
#   sub_scaler  — fit on subset train, used during Optuna
#   full_scaler — fit on full train,   used for final model
# Each scaler is fitted on its respective training partition
# only, preventing leakage from val or test statistics.
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 3: FEATURE SCALING")
print("=" * 60)

sub_scaler  = StandardScaler()
X_sub_tr    = sub_scaler.fit_transform(X_sub_tr_raw)
X_sub_vl    = sub_scaler.transform(X_sub_vl_raw)

full_scaler = StandardScaler()
X_tr        = full_scaler.fit_transform(X_tr_raw)
X_vl        = full_scaler.transform(X_vl_raw)
X_te        = full_scaler.transform(X_te_raw)

print(f"  Subset scaler — mean: {X_sub_tr.mean():.6f}  std: {X_sub_tr.std():.6f}")
print(f"  Full   scaler — mean: {X_tr.mean():.6f}  std: {X_tr.std():.6f}")


# ─────────────────────────────────────────────────────────────
# BLOCK 4: IMBALANCE HANDLING — pos_weight
#
# pos_weight is computed separately for:
#   - Subset train  : used during Optuna trials
#   - Full train    : used for final model and CV folds
# No undersampling. No synthetic oversampling.
# Temporal sequence preserved. shuffle=False throughout.
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 4: IMBALANCE HANDLING")
print("=" * 60)
print("  Strategy : pos_weight in BCEWithLogitsLoss only.")
print("             All samples retained. Temporal order preserved.")
print("             shuffle=False in all DataLoaders.")

sub_neg        = (y_sub_tr == 0).sum()
sub_pos        = (y_sub_tr == 1).sum()
SUB_POS_WEIGHT = torch.tensor([sub_neg / sub_pos], dtype=torch.float32).to(DEVICE)

full_neg        = (y_tr == 0).sum()
full_pos        = (y_tr == 1).sum()
FULL_POS_WEIGHT = torch.tensor([full_neg / full_pos], dtype=torch.float32).to(DEVICE)

print(f"  Subset train — Normal: {int(sub_neg):,} | Violated: {int(sub_pos):,} | pos_weight: {SUB_POS_WEIGHT.item():.4f}")
print(f"  Full   train — Normal: {int(full_neg):,} | Violated: {int(full_pos):,} | pos_weight: {FULL_POS_WEIGHT.item():.4f}")


# ─────────────────────────────────────────────────────────────
# BLOCK 5: MLP ARCHITECTURE
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 5: MLP ARCHITECTURE")
print("=" * 60)
print("  Per layer : Linear -> BatchNorm -> Activation -> Dropout")
print("  Output    : Raw logit — sigmoid fused in BCEWithLogitsLoss")
print("  Init      : Xavier uniform")

INPUT_DIM = X_sub_tr.shape[1]


class MLPClassifier(nn.Module):

    def __init__(self, input_dim, hidden_dims, dropout_rate, activation):
        super().__init__()
        act_map = {'relu': nn.ReLU, 'tanh': nn.Tanh}
        layers  = []
        in_dim  = input_dim
        for out_dim in hidden_dims:
            layers += [
                nn.Linear(in_dim, out_dim),
                nn.BatchNorm1d(out_dim),
                act_map[activation](),
                nn.Dropout(p=dropout_rate)
            ]
            in_dim = out_dim
        layers.append(nn.Linear(in_dim, 1))
        self.net = nn.Sequential(*layers)
        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                nn.init.zeros_(m.bias)

    def forward(self, x):
        return self.net(x).squeeze(1)

print("  Architecture class defined.")


# ─────────────────────────────────────────────────────────────
# BLOCK 6: TRAINING FUNCTION
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 6: TRAINING FUNCTION")
print("=" * 60)
print("  Optimizer  : AdamW")
print("  Scheduler  : ReduceLROnPlateau (val AUPRC)")
print("  shuffle=False — temporal order always preserved")
print("  Imbalance  : pos_weight only")
print("  Grad clip  : L2 norm 1.0")


def train_model(model, X_tr_in, y_tr_in, X_vl_in, y_vl_in,
                lr, batch_size, epochs, weight_decay,
                patience=15, label="Training",
                pos_weight=None):

    if pos_weight is None:
        pos_weight = FULL_POS_WEIGHT

    criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=0.5, patience=5, min_lr=1e-6
    )

    loader = DataLoader(
        TensorDataset(
            torch.tensor(X_tr_in, dtype=torch.float32),
            torch.tensor(y_tr_in, dtype=torch.float32)
        ),
        batch_size=batch_size,
        shuffle=False
    )

    X_vl_t = torch.tensor(X_vl_in, dtype=torch.float32).to(DEVICE)

    history = {
        'train_loss': [], 'val_loss'  : [],
        'val_auprc' : [], 'val_auroc' : [],
        'grad_norm' : [], 'lr'        : []
    }

    best_val_auprc = 0.0
    best_weights   = None
    no_improve     = 0
    pbar           = tqdm(range(epochs), desc=label, leave=True)

    for epoch in pbar:
        model.train()
        total_loss  = 0.0
        total_gnorm = 0.0
        n_batches   = 0

        for Xb, yb in loader:
            Xb, yb = Xb.to(DEVICE), yb.to(DEVICE)
            optimizer.zero_grad()
            loss = criterion(model(Xb), yb)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            gnorm = sum(
                p.grad.norm().item() ** 2
                for p in model.parameters() if p.grad is not None
            ) ** 0.5
            optimizer.step()
            total_loss  += loss.item()
            total_gnorm += gnorm
            n_batches   += 1

        model.eval()
        with torch.no_grad():
            vl_logits = model(X_vl_t).cpu().numpy()
            vl_probs  = 1 / (1 + np.exp(-vl_logits))
            vl_loss   = criterion(
                torch.tensor(vl_logits).to(DEVICE),
                torch.tensor(y_vl_in, dtype=torch.float32).to(DEVICE)
            ).item()

        val_auprc = average_precision_score(y_vl_in, vl_probs)
        val_auroc = roc_auc_score(y_vl_in, vl_probs)
        cur_lr    = optimizer.param_groups[0]['lr']

        history['train_loss'].append(total_loss / n_batches)
        history['val_loss'  ].append(vl_loss)
        history['val_auprc' ].append(val_auprc)
        history['val_auroc' ].append(val_auroc)
        history['grad_norm' ].append(total_gnorm / n_batches)
        history['lr'        ].append(cur_lr)

        scheduler.step(val_auprc)
        pbar.set_postfix({
            'loss'     : f"{total_loss/n_batches:.4f}",
            'val_auprc': f"{val_auprc:.4f}",
            'gnorm'    : f"{total_gnorm/n_batches:.3f}",
            'lr'       : f"{cur_lr:.6f}",
            'patience' : f"{no_improve}/{patience}"
        })

        if val_auprc > best_val_auprc:
            best_val_auprc = val_auprc
            best_weights   = {k: v.clone() for k, v in model.state_dict().items()}
            no_improve     = 0
        else:
            no_improve += 1

        if no_improve >= patience:
            tqdm.write(
                f"  Early stopping — epoch {epoch+1} | "
                f"Best Val AUPRC: {best_val_auprc:.4f}"
            )
            break

    model.load_state_dict(best_weights)
    return model, history, best_val_auprc

print("  Training function defined.")


# ─────────────────────────────────────────────────────────────
# BLOCK 7: OPTUNA — 25 TRIALS ON SUBSET
#
# All 25 trials run on the temporal subset (train/val).
# Val is strictly later in time than train.
# SUB_POS_WEIGHT applied in every trial.
#
# Search space:
#   hidden_layers [1, 3]        — network depth
#   neurons       [64,128,256]  — width per layer
#   learning_rate [1e-4, 1e-2]  — AdamW LR, log scale
#   activation    [relu, tanh]  — nonlinearity
#   epochs        [50, 200]     — upper bound; early stopping
#   dropout       [0.1, 0.4]    — regularisation
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 7: OPTUNA — 25 TRIALS ON SUBSET")
print("=" * 60)
print("  Sampler    : TPE")
print("  Objective  : Maximize subset Val AUPRC")
print("  Data       : subset train / val (temporal, no shuffle)")
print("  Parameters : hidden_layers, neurons, learning_rate,")
print("               activation, epochs, dropout")
print("-" * 60)

all_trial_results = []


def objective(trial):
    n_layers   = trial.suggest_int('hidden_layers', 1, 3)
    neurons    = trial.suggest_categorical('neurons', [64, 128, 256])
    lr         = trial.suggest_float('learning_rate', 1e-4, 1e-2, log=True)
    activation = trial.suggest_categorical('activation', ['relu', 'tanh'])
    epochs     = trial.suggest_int('epochs', 50, 200)
    dropout    = trial.suggest_float('dropout', 0.1, 0.4)

    hidden_dims = [neurons] * n_layers

    print(f"\n  Trial {trial.number+1:02d}/25 | "
          f"Layers={n_layers} Neurons={neurons} | "
          f"act={activation} | lr={lr:.5f} | "
          f"dropout={dropout:.2f} | epochs={epochs}")

    model = MLPClassifier(INPUT_DIM, hidden_dims, dropout, activation).to(DEVICE)

    t0 = time.time()
    _, history, val_auprc = train_model(
        model,
        X_sub_tr, y_sub_tr,
        X_sub_vl, y_sub_vl,
        lr           = lr,
        batch_size   = 256,
        epochs       = epochs,
        weight_decay = 1e-4,
        patience     = 10,
        label        = f"Trial {trial.number+1:02d}/25",
        pos_weight   = SUB_POS_WEIGHT
    )
    elapsed = time.time() - t0

    print(f"  Trial {trial.number+1:02d}/25 | "
          f"Val AUPRC: {val_auprc:.4f} | "
          f"Epochs run: {len(history['train_loss'])} | "
          f"Time: {elapsed:.0f}s")

    trial_name = (
        f"mlp_trial{trial.number:02d}"
        f"_L{n_layers}_N{neurons}"
        f"_{activation}_dr{dropout:.2f}"
    )
    torch.save(model.state_dict(), SAVE_DIR + trial_name + '.pt')
    joblib.dump(history,           SAVE_DIR + trial_name + '_history.pkl')

    all_trial_results.append({
        'trial'           : trial.number,
        'hidden_layers'   : n_layers,
        'neurons'         : neurons,
        'architecture'    : str(hidden_dims),
        'activation'      : activation,
        'learning_rate'   : round(lr, 6),
        'dropout'         : round(dropout, 3),
        'epochs_allowed'  : epochs,
        'epochs_run'      : len(history['train_loss']),
        'val_auprc'       : round(val_auprc, 4),
        'final_train_loss': round(history['train_loss'][-1], 6),
        'final_val_loss'  : round(history['val_loss'][-1], 6),
        'trial_time_sec'  : round(elapsed, 1),
        'model_file'      : trial_name + '.pt',
        'history_file'    : trial_name + '_history.pkl'
    })

    return val_auprc


study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=25)

print("\n" + "=" * 60)
print("OPTUNA COMPLETE")
print(f"  Best Val AUPRC : {study.best_value:.4f}")
print(f"  Best Trial     : {study.best_trial.number + 1}")
print("  Best Params    :")
for k, v in study.best_params.items():
    print(f"    {k:<18}: {v}")
print("=" * 60)


# ─────────────────────────────────────────────────────────────
# BLOCK 8: TRIAL RESULTS TABLE
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 8: TRIAL RESULTS")
print("=" * 60)

results_df = pd.DataFrame(all_trial_results).sort_values(
    'val_auprc', ascending=False
).reset_index(drop=True)

results_df.to_csv(SAVE_DIR + 'all_trial_results.csv', index=False)

print("\nALL TRIALS RANKED BY VAL AUPRC:")
print("-" * 85)
print(results_df[[
    'trial', 'hidden_layers', 'neurons', 'activation',
    'learning_rate', 'dropout', 'epochs_run', 'val_auprc'
]].to_string(index=False))

with open(SAVE_DIR + 'trial_log.txt', 'w') as f:
    f.write("MLP HYPERPARAMETER TUNING LOG\n")
    f.write("Dataset    : Alibaba MS Trace v2022\n")
    f.write("Framework  : PyTorch\n")
    f.write("Optimizer  : AdamW\n")
    f.write("Scheduler  : ReduceLROnPlateau\n")
    f.write("Imbalance  : pos_weight only — no undersampling\n")
    f.write("DataLoader : shuffle=False throughout\n")
    f.write("Optuna data: subset (middle 30%, idx 40%->70%)\n")
    f.write("Final train: full dataset\n")
    f.write("Trials     : 25 | Metric : Val AUPRC\n")
    f.write("=" * 70 + "\n\n")
    for _, row in results_df.iterrows():
        f.write(f"TRIAL {int(row['trial']):02d} | Val AUPRC : {row['val_auprc']}\n")
        f.write(f"  Hidden Layers : {row['hidden_layers']}\n")
        f.write(f"  Neurons       : {row['neurons']}\n")
        f.write(f"  Activation    : {row['activation']}\n")
        f.write(f"  Learning Rate : {row['learning_rate']}\n")
        f.write(f"  Dropout       : {row['dropout']}\n")
        f.write(f"  Epochs Run    : {row['epochs_run']}\n")
        f.write(f"  Train Loss    : {row['final_train_loss']}\n")
        f.write(f"  Val Loss      : {row['final_val_loss']}\n")
        f.write(f"  Time (sec)    : {row['trial_time_sec']}\n")
        f.write("-" * 50 + "\n")
    f.write(f"\nBEST TRIAL : {results_df.iloc[0]['trial']}\n")
    f.write(f"BEST AUPRC : {results_df.iloc[0]['val_auprc']}\n")

joblib.dump(study, SAVE_DIR + 'optuna_study.pkl')
print("  all_trial_results.csv : Saved")
print("  trial_log.txt         : Saved")
print("  optuna_study.pkl      : Saved")


# ─────────────────────────────────────────────────────────────
# BLOCK 9: RETRAIN BEST MODEL ON FULL TRAINING DATA
#
# Hyperparameters from Optuna subset search applied to a
# model trained on the full training partition with
# full_scaler and FULL_POS_WEIGHT. This is the model used
# for all reported evaluation metrics.
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 9: RETRAINING BEST MODEL ON FULL TRAINING DATA")
print("=" * 60)

bp          = study.best_params
best_hidden = [bp['neurons']] * bp['hidden_layers']

print(f"  Architecture  : {best_hidden}")
print(f"  Activation    : {bp['activation']}")
print(f"  Learning Rate : {bp['learning_rate']}")
print(f"  Dropout       : {bp['dropout']}")
print(f"  Max Epochs    : {bp['epochs']} | Patience : 15")
print(f"  Training data : full train ({len(X_tr):,} samples)")
print(f"  pos_weight    : {FULL_POS_WEIGHT.item():.4f}")

best_model = MLPClassifier(
    INPUT_DIM, best_hidden, bp['dropout'], bp['activation']
).to(DEVICE)

best_model, best_history, best_val_auprc = train_model(
    best_model,
    X_tr, y_tr,
    X_vl, y_vl,
    lr           = bp['learning_rate'],
    batch_size   = 256,
    epochs       = bp['epochs'],
    weight_decay = 1e-4,
    patience     = 15,
    label        = "Best model — full data",
    pos_weight   = FULL_POS_WEIGHT
)

torch.save(best_model.state_dict(), SAVE_DIR + 'mlp_best_model.pt')
joblib.dump(best_history,           SAVE_DIR + 'mlp_best_history.pkl')
joblib.dump(full_scaler,            SAVE_DIR + 'scaler.pkl')
joblib.dump(bp,                     SAVE_DIR + 'best_params.pkl')

print(f"  Best Val AUPRC   : {best_val_auprc:.4f}")
print(f"  Epochs trained   : {len(best_history['train_loss'])}")
print("  mlp_best_model.pt    : Saved")
print("  mlp_best_history.pkl : Saved")
print("  scaler.pkl           : Saved")
print("  best_params.pkl      : Saved")


# ─────────────────────────────────────────────────────────────
# BLOCK 10: TEMPORAL K-FOLD CV (K=5) ON FULL TRAINING DATA
#
# Cross-validation is performed on the full training partition.
# For fold i (i = 1..K-1):
#   train = windows in folds 0..i-1   (strictly the past)
#   val   = windows in fold i          (strictly the future)
# pos_weight recomputed per fold from that fold's train
# partition — violation rate varies across time periods.
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 10: TEMPORAL K-FOLD CV (K=5)")
print("=" * 60)
print("  Data   : full training partition")
print("  Method : train=past folds, val=next fold (no shuffle)")
print("  pos_weight recomputed per fold")
print("-" * 60)

K            = 5
n_tr_full    = len(X_tr)
fold_size    = n_tr_full // K
fold_auprcs  = []
fold_aurocs  = []
fold_details = []

for fold_num in range(1, K):
    tr_end   = fold_num * fold_size
    vl_start = fold_num * fold_size
    vl_end   = min((fold_num + 1) * fold_size, n_tr_full)

    X_fold_tr = X_tr[:tr_end];         y_fold_tr = y_tr[:tr_end]
    X_fold_vl = X_tr[vl_start:vl_end]; y_fold_vl = y_tr[vl_start:vl_end]

    fold_neg = (y_fold_tr == 0).sum()
    fold_pos = (y_fold_tr == 1).sum()
    if fold_pos == 0:
        print(f"  Fold {fold_num}/{K-1} skipped — no violations in train partition")
        continue

    fold_pw = torch.tensor([fold_neg / fold_pos], dtype=torch.float32).to(DEVICE)
    vl_pos  = (y_fold_vl == 1).sum()

    print(f"\n  Fold {fold_num}/{K-1} | "
          f"train 0..{tr_end-1} ({tr_end:,}) | "
          f"val {vl_start}..{vl_end-1} ({vl_end-vl_start:,}) | "
          f"viol={int(vl_pos)} | pw={fold_pw.item():.2f}")

    fold_model = MLPClassifier(
        INPUT_DIM, best_hidden, bp['dropout'], bp['activation']
    ).to(DEVICE)

    fold_model, _, _ = train_model(
        fold_model,
        X_fold_tr, y_fold_tr,
        X_fold_vl, y_fold_vl,
        lr           = bp['learning_rate'],
        batch_size   = 256,
        epochs       = bp['epochs'],
        weight_decay = 1e-4,
        patience     = 10,
        label        = f"Fold {fold_num}/{K-1}",
        pos_weight   = fold_pw
    )

    fold_model.eval()
    with torch.no_grad():
        fold_probs = 1 / (1 + np.exp(
            -fold_model(
                torch.tensor(X_fold_vl, dtype=torch.float32).to(DEVICE)
            ).cpu().numpy()
        ))

    fa = average_precision_score(y_fold_vl, fold_probs)
    fr = roc_auc_score(y_fold_vl, fold_probs)
    fold_auprcs.append(fa)
    fold_aurocs.append(fr)
    fold_details.append({
        'fold'            : fold_num,
        'train_samples'   : tr_end,
        'val_samples'     : vl_end - vl_start,
        'val_viol_pct'    : round(float(vl_pos / len(y_fold_vl) * 100), 2),
        'fold_pos_weight' : round(fold_pw.item(), 3),
        'auprc'           : round(fa, 4),
        'auroc'           : round(fr, 4)
    })
    print(f"  Fold {fold_num}/{K-1} | AUPRC={fa:.4f} | AUROC={fr:.4f}")

print(f"\n  Temporal CV AUPRC : {[round(x,4) for x in fold_auprcs]}")
if fold_auprcs:
    print(f"  Mean              : {np.mean(fold_auprcs):.4f} +/- {np.std(fold_auprcs):.4f}")

joblib.dump({
    'fold_auprcs'  : fold_auprcs,
    'fold_aurocs'  : fold_aurocs,
    'mean_auprc'   : float(np.mean(fold_auprcs)) if fold_auprcs else None,
    'std_auprc'    : float(np.std(fold_auprcs))  if fold_auprcs else None,
    'fold_details' : fold_details
}, SAVE_DIR + 'kfold_results.pkl')
print("  kfold_results.pkl : Saved")


# ─────────────────────────────────────────────────────────────
# BLOCK 11: THRESHOLD OPTIMISATION
#
# Optimised on the full validation set by maximising F2.
# F2 weights recall twice over precision — missing a true
# SLA violation carries higher cost than a false alarm.
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 11: THRESHOLD OPTIMISATION")
print("=" * 60)
print("  Metric : F2 Score on full validation set")

best_model.eval()
with torch.no_grad():
    val_probs = 1 / (1 + np.exp(
        -best_model(
            torch.tensor(X_vl, dtype=torch.float32).to(DEVICE)
        ).cpu().numpy()
    ))

precision_vals, recall_vals, thresholds = precision_recall_curve(y_vl, val_probs)

f2_scores = np.array([
    fbeta_score(y_vl, (val_probs >= t).astype(int), beta=2, zero_division=0)
    for t in thresholds
])
f1_scores = np.array([
    f1_score(y_vl, (val_probs >= t).astype(int), zero_division=0)
    for t in thresholds
])

best_thresh    = thresholds[np.argmax(f2_scores)]
best_f1_thresh = thresholds[np.argmax(f1_scores)]

print(f"  Optimal threshold (F2) : {best_thresh:.4f}")
print(f"  Optimal threshold (F1) : {best_f1_thresh:.4f}")
print(f"  Val F2 at optimal      : {f2_scores.max():.4f}")
print(f"  Val F1 at optimal      : {f1_scores.max():.4f}")


# ─────────────────────────────────────────────────────────────
# BLOCK 12: FINAL TEST SET EVALUATION
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 12: FINAL TEST SET EVALUATION")
print("=" * 60)

best_model.eval()
with torch.no_grad():
    test_probs = 1 / (1 + np.exp(
        -best_model(
            torch.tensor(X_te, dtype=torch.float32).to(DEVICE)
        ).cpu().numpy()
    ))

test_preds_05  = (test_probs >= 0.50).astype(int)
test_preds_opt = (test_probs >= best_thresh).astype(int)

test_auprc  = average_precision_score(y_te, test_probs)
test_auroc  = roc_auc_score(y_te, test_probs)
prec_05     = precision_score(y_te, test_preds_05,  zero_division=0)
rec_05      = recall_score(y_te, test_preds_05,     zero_division=0)
prec_opt    = precision_score(y_te, test_preds_opt, zero_division=0)
rec_opt     = recall_score(y_te, test_preds_opt,    zero_division=0)
f1_05       = f1_score(y_te, test_preds_05,  zero_division=0)
f2_05       = fbeta_score(y_te, test_preds_05,  beta=2, zero_division=0)
f1_opt      = f1_score(y_te, test_preds_opt, zero_division=0)
f2_opt      = fbeta_score(y_te, test_preds_opt, beta=2, zero_division=0)
acc_05      = (test_preds_05  == y_te).mean()
acc_opt     = (test_preds_opt == y_te).mean()

print(f"\n  {'Metric':<22} {'Threshold=0.50':>16} {'Optimal ({:.3f})'.format(best_thresh):>18}")
print(f"  {'_'*58}")
print(f"  {'AUPRC':<22} {'N/A':>16} {test_auprc:>18.4f}")
print(f"  {'AUROC':<22} {'N/A':>16} {test_auroc:>18.4f}")
print(f"  {'Precision':<22} {prec_05:>16.4f} {prec_opt:>18.4f}")
print(f"  {'Recall':<22} {rec_05:>16.4f} {rec_opt:>18.4f}")
print(f"  {'F1 Score':<22} {f1_05:>16.4f} {f1_opt:>18.4f}")
print(f"  {'F2 Score':<22} {f2_05:>16.4f} {f2_opt:>18.4f}")
print(f"  {'Accuracy':<22} {acc_05:>16.4f} {acc_opt:>18.4f}")
print(f"\n  Classification Report at Optimal Threshold:")
print(classification_report(y_te, test_preds_opt, zero_division=0))


# ─────────────────────────────────────────────────────────────
# BLOCK 13: VISUALIZATION
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 13: GENERATING PLOTS")
print("=" * 60)

epochs_ax = range(1, len(best_history['train_loss']) + 1)
best_ep   = int(np.argmax(best_history['val_auprc'])) + 1
best_row  = results_df.iloc[0]

fig = plt.figure(figsize=(20, 15))
fig.suptitle(
    'MLP Classifier — Training Analysis\n'
    'Alibaba MS Trace v2022 | pos_weight | Full Data Training | Temporal CV K=5',
    fontsize=14, y=0.99
)
gs = gridspec.GridSpec(3, 4, figure=fig, hspace=0.5, wspace=0.38)

ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(epochs_ax, best_history['train_loss'], color='steelblue', lw=2, label='Train')
ax1.plot(epochs_ax, best_history['val_loss'],   color='darkorange', lw=2, label='Val')
ax1.axvline(x=best_ep, color='crimson', ls='--', label=f'Best ep {best_ep}')
ax1.set_xlabel('Epoch'); ax1.set_ylabel('BCE Loss')
ax1.set_title('Train vs Validation Loss')
ax1.legend(fontsize=8); ax1.grid(True, alpha=0.3)

ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(epochs_ax, best_history['val_auprc'], color='seagreen', lw=2)
ax2.axvline(x=best_ep, color='crimson', ls='--', label=f'Best={max(best_history["val_auprc"]):.4f}')
ax2.set_xlabel('Epoch'); ax2.set_ylabel('AUPRC')
ax2.set_title('Validation AUPRC per Epoch')
ax2.legend(fontsize=8); ax2.grid(True, alpha=0.3)

ax3 = fig.add_subplot(gs[0, 2])
ax3.plot(epochs_ax, best_history['grad_norm'], color='purple', lw=1.5)
ax3.set_xlabel('Epoch'); ax3.set_ylabel('Gradient L2 Norm')
ax3.set_title('Gradient Norm per Epoch')
ax3.grid(True, alpha=0.3)

ax4 = fig.add_subplot(gs[0, 3])
ax4.plot(epochs_ax, best_history['lr'], color='steelblue', lw=2)
ax4.set_xlabel('Epoch'); ax4.set_ylabel('Learning Rate')
ax4.set_title('LR Schedule (ReduceLROnPlateau)')
ax4.grid(True, alpha=0.3)

ax5 = fig.add_subplot(gs[1, 0])
cm = confusion_matrix(y_te, test_preds_opt)
ax5.imshow(cm, cmap='Blues')
ax5.set_xticks([0,1]); ax5.set_yticks([0,1])
ax5.set_xticklabels(['Normal', 'Violated'])
ax5.set_yticklabels(['Normal', 'Violated'])
ax5.set_xlabel('Predicted'); ax5.set_ylabel('Actual')
ax5.set_title(f'Confusion Matrix\nThreshold = {best_thresh:.3f}')
for i in range(2):
    for j in range(2):
        ax5.text(j, i, str(cm[i,j]), ha='center', va='center', fontsize=13)

ax6 = fig.add_subplot(gs[1, 1])
prec_c, rec_c, _ = precision_recall_curve(y_te, test_probs)
ax6.plot(rec_c, prec_c, color='steelblue', lw=2)
ax6.scatter(recall_vals[np.argmax(f2_scores)],
            precision_vals[np.argmax(f2_scores)],
            color='crimson', s=150, zorder=5, label='Optimal point')
ax6.set_xlabel('Recall'); ax6.set_ylabel('Precision')
ax6.set_title(f'Precision-Recall Curve\nAUPRC = {test_auprc:.4f}')
ax6.legend(fontsize=8); ax6.grid(True, alpha=0.3)

ax7 = fig.add_subplot(gs[1, 2])
ax7.plot(thresholds, f1_scores, color='darkorange', lw=2, label='F1')
ax7.plot(thresholds, f2_scores, color='steelblue',  lw=2, label='F2')
ax7.axvline(x=best_thresh, color='crimson', ls='--', label=f'Optimal = {best_thresh:.3f}')
ax7.set_xlabel('Threshold'); ax7.set_ylabel('Score')
ax7.set_title('F1 and F2 vs Decision Threshold')
ax7.legend(fontsize=8); ax7.grid(True, alpha=0.3)

ax8 = fig.add_subplot(gs[1, 3])
fold_labels = [f'Fold {d["fold"]}\n({d["train_samples"]:,} tr)' for d in fold_details]
ax8.bar(fold_labels, fold_auprcs, color='steelblue', edgecolor='white')
if fold_auprcs:
    ax8.axhline(y=np.mean(fold_auprcs), color='crimson', ls='--', label=f'Mean = {np.mean(fold_auprcs):.4f}')
ax8.set_ylabel('AUPRC')
ax8.set_title('Temporal CV AUPRC per Fold\n(train=past, val=future)')
ax8.legend(fontsize=8); ax8.grid(True, alpha=0.3, axis='y')

ax9 = fig.add_subplot(gs[2, 0])
lg = results_df.groupby('hidden_layers')['val_auprc'].mean().reset_index()
ax9.bar(lg['hidden_layers'].astype(str), lg['val_auprc'],
        color='steelblue', edgecolor='white')
ax9.set_xlabel('Hidden Layers'); ax9.set_ylabel('Mean Val AUPRC')
ax9.set_title('Effect of Network Depth')
ax9.grid(True, alpha=0.3, axis='y')

ax10 = fig.add_subplot(gs[2, 1])
ng = results_df.groupby('neurons')['val_auprc'].mean().reset_index()
ax10.bar(ng['neurons'].astype(str), ng['val_auprc'],
         color='steelblue', edgecolor='white')
ax10.set_xlabel('Neurons per Layer'); ax10.set_ylabel('Mean Val AUPRC')
ax10.set_title('Effect of Neuron Count')
ax10.grid(True, alpha=0.3, axis='y')

ax11 = fig.add_subplot(gs[2, 2])
ag = results_df.groupby('activation')['val_auprc'].mean().reset_index()
ax11.bar(ag['activation'], ag['val_auprc'],
         color='steelblue', edgecolor='white')
ax11.set_xlabel('Activation'); ax11.set_ylabel('Mean Val AUPRC')
ax11.set_title('Effect of Activation Function')
ax11.grid(True, alpha=0.3, axis='y')

ax12 = fig.add_subplot(gs[2, 3])
ax12.scatter(results_df['learning_rate'], results_df['val_auprc'],
             alpha=0.7, color='steelblue', s=60)
ax12.scatter(best_row['learning_rate'], best_row['val_auprc'],
             color='crimson', s=150, zorder=5, marker='*', label=f"Best: {best_row['learning_rate']:.5f}")
ax12.set_xscale('log')
ax12.set_xlabel('Learning Rate (log)'); ax12.set_ylabel('Val AUPRC')
ax12.set_title('Effect of Learning Rate')
ax12.legend(fontsize=8); ax12.grid(True, alpha=0.3)

plt.savefig(SAVE_DIR + 'mlp_full_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("  mlp_full_analysis.png : Saved")


# ─────────────────────────────────────────────────────────────
# BLOCK 14: FINAL SUMMARY
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 60)
print("BLOCK 14: FINAL SUMMARY")
print("=" * 60)

bp_serializable = {
    k: (int(v)   if isinstance(v, (int,   np.integer))
        else float(v) if isinstance(v, (float, np.floating))
        else v)
    for k, v in bp.items()
}

prec_normal_val  = float(precision_score(y_te, test_preds_opt, pos_label=0, average='binary', zero_division=0))
rec_normal_val   = float(recall_score(y_te,    test_preds_opt, pos_label=0, average='binary', zero_division=0))
prec_overall_val = float(precision_score(y_te, test_preds_opt, average='weighted', zero_division=0))
rec_overall_val  = float(recall_score(y_te,    test_preds_opt, average='weighted', zero_division=0))

summary = {
    'model'                   : 'MLP — PyTorch',
    'optuna_data'             : f'Temporal subset — middle 30% (idx {sub_start}->{sub_end})',
    'final_train_data'        : f'Full train ({len(X_tr):,} samples)',
    'imbalance_strategy'      : 'pos_weight in BCEWithLogitsLoss — no undersampling',
    'temporal_integrity'      : 'shuffle=False in all DataLoaders',
    'cv_method'               : 'Temporal K-Fold K=5 — train=past, val=future',
    'optimizer'               : 'AdamW',
    'scheduler'               : 'ReduceLROnPlateau',
    'loss_function'           : 'BCEWithLogitsLoss with pos_weight',
    'regularization'          : 'Dropout + L2 weight decay + gradient clipping',
    'initialization'          : 'Xavier uniform',
    'best_architecture'       : str(best_hidden),
    'best_params'             : bp_serializable,
    'optimal_threshold'       : round(float(best_thresh), 4),
    'kfold_mean_auprc'        : round(float(np.mean(fold_auprcs)), 4) if fold_auprcs else None,
    'kfold_std_auprc'         : round(float(np.std(fold_auprcs)),  4) if fold_auprcs else None,
    'val_auprc'               : round(float(best_val_auprc), 4),
    'test_auprc'              : round(float(test_auprc), 4),
    'test_auroc'              : round(float(test_auroc), 4),
    'test_precision_normal'   : round(prec_normal_val, 4),
    'test_recall_normal'      : round(rec_normal_val, 4),
    'test_precision_violated' : round(float(prec_opt), 4),
    'test_recall_violated'    : round(float(rec_opt), 4),
    'test_precision_overall'  : round(prec_overall_val, 4),
    'test_recall_overall'     : round(rec_overall_val, 4),
    'test_f1'                 : round(float(f1_opt), 4),
    'test_f2'                 : round(float(f2_opt), 4),
    'test_accuracy'           : round(float(acc_opt), 4),
    'total_trials'            : len(results_df)
}

with open(SAVE_DIR + 'mlp_best_summary.json', 'w') as f:
    json.dump(summary, f, indent=4)

print(f"\n  SAVE STRUCTURE:")
print(f"  {SAVE_DIR}")
print(f"    mlp_best_model.pt")
print(f"    mlp_best_history.pkl")
print(f"    scaler.pkl")
print(f"    best_params.pkl")
print(f"    all_trial_results.csv")
print(f"    trial_log.txt")
print(f"    optuna_study.pkl")
print(f"    kfold_results.pkl")
print(f"    mlp_best_summary.json")
print(f"    mlp_full_analysis.png")
print(f"    mlp_trial00.pt ... mlp_trial24.pt")

print(f"\n  MODEL CONFIGURATION:")
print(f"    Hyperparameter search  : Optuna TPE, 25 trials on temporal subset")
print(f"    Final training data    : Full train ({len(X_tr):,} samples)")
print(f"    pos_weight (full)      : {FULL_POS_WEIGHT.item():.4f}")
print(f"    Cross-validation       : Temporal K-Fold K=5 on full train")
print(f"    Threshold optimisation : F2-optimal on full validation set")

print(f"\n  BEST HYPERPARAMETERS:")
for k, v in bp_serializable.items():
    print(f"    {k:<22}: {v}")

print(f"\n  VALIDATION METRICS (full validation set):")
print(f"    Val AUPRC              : {best_val_auprc:.4f}")

print(f"\n  FINAL TEST METRICS (threshold = {best_thresh:.4f}):")
print(f"    Test AUPRC             : {test_auprc:.4f}")
print(f"    Test AUROC             : {test_auroc:.4f}")
print(f"    F1 Score               : {f1_opt:.4f}")
print(f"    F2 Score               : {f2_opt:.4f}")
print(f"    Accuracy               : {acc_opt:.4f}")

print(f"\n  PRECISION (class-wise and overall):")
print(f"    Normal    (class 0)    : {prec_normal_val:.4f}")
print(f"    Violated  (class 1)    : {prec_opt:.4f}")
print(f"    Weighted overall       : {prec_overall_val:.4f}")

print(f"\n  RECALL (class-wise and overall):")
print(f"    Normal    (class 0)    : {rec_normal_val:.4f}")
print(f"    Violated  (class 1)    : {rec_opt:.4f}")
print(f"    Weighted overall       : {rec_overall_val:.4f}")

print(f"\n  CROSS-VALIDATION (temporal K-fold, K=5):")
if fold_auprcs:
    print(f"    Mean AUPRC             : {np.mean(fold_auprcs):.4f} +/- {np.std(fold_auprcs):.4f}")
    print(f"    Per-fold AUPRC         : {[round(x,4) for x in fold_auprcs]}")
else:
    print(f"    No folds completed — all folds had zero violations in train partition")

print("\n" + "=" * 60)
print("ALL BLOCKS COMPLETE.")
print("=" * 60)
```
